Para começarmos a mergulhar na solução do problema, inicialmente devevemos relembrar o que é DNA, RNA e suas principais diferenças, além de conhecermos o conceito de alelos  e por fim o conceito sobre o que é bioinformática, sendo assim a seguir apresentaremos cada um desses termos em tópicos para facilitar o seu entendimento.

## **&#9776; DNA**

O DNA (ácido desoxirribonucleico) é um tipo de ácido nucleico que possui destaque por armazenar a informação genética da grande maioria dos seres vivos. Essa molécula é formada por nucleotídeos e apresenta, geralmente, a forma de uma dupla-hélice. 

![Picture](https://s1.static.brasilescola.uol.com.br/be/conteudo/images/1-dna.jpg "DNA")

## **Composição do DNA**

O DNA é composto por nucleotídeos, os quais são compostos por três partes:

- Um carboidrato de cinco carbonos (pentose)

- Uma base nitrogenada

- Um ou mais grupos fosfato.

## **Estrutura do DNA**

O DNA é formado por duas cadeias de polinucleotídeos (fita), que são constituídas por vários nucleotídeos. Os nucleotídeos são unidos uns aos outros por ligações denominadas fosfodiéster (grupo fosfato ligando dois açúcares de dois nucleotídeos). Nessas ligações, um grupo fosfato conecta o carbono 3’ de um açúcar ao carbono 5’ do próximo açúcar.

Essa junção dos nucleotídeos forma um padrão típico de repetição de unidade de açúcar-fosfato, que forma a cadeia principal. A essa cadeia principal estão ligadas as bases nitrogenadas.

![Picture](https://s3.static.brasilescola.uol.com.br/img/2019/03/estrutura-do-dna.jpg "Estrutura do DNA")


## **Função do DNA**

O DNA é uma molécula extremamente importante para os seres vivos. São funções do DNA:

- Armazenar e transmitir as informações genéticas.

- Funcionar como molde para a síntese da molécula de RNA. O DNA, portanto, é fundamental para a síntese de proteínas, uma vez que contém as informações que comandam a síntese de RNA, e o RNA coordena a produção desses polipeptídeos (DNA → RNA → Proteína).


## **&#9776; RNA**

Assim como o DNA, o RNA (ácido ribonucleico) é um ácido nucleico. Essa molécula é essencial na síntese de proteínas, já que ela funciona como uma intermediadora capaz de expressar as informações presentes no DNA. A molécula de RNA é formada a partir da molécula de DNA em um processo chamado de transcrição.

![Picture](https://s1.static.brasilescola.uol.com.br/be/conteudo/images/o-rna-pode-ser-encontrado-em-tres-formas-rnam-rnat-rnar-5970c73938023.jpg "RNA")

O RNA pode ser encontrado em três formas: RNAm, RNAt e RNAr

## **Tipos de RNA**

-**RNA mensageiro (RNAm)**: Esse tipo de RNA codifica as proteínas, uma vez que porta as informações do DNA. Simplificadamente, é o RNAm que é traduzido no processo de formação das proteínas, chamado, portanto, de tradução.

-**RNA transportador (RNAt)**: O RNA transportador, ou RNA de transferência, é o responsável por transportar os aminoácidos que formarão a nova proteína. Esse RNA identifica a sequência de três nucleotídios que codificam um aminoácido (códon) e garante que o aminoácido correspondente àquela informação seja adicionado à cadeia em formação.

-**RNA ribossomal (RNAr)**: Esse RNA forma os ribossomos, que são os locais onde ocorrem a síntese de proteínas. É importante destacar que 80% do RNA presente na célula é desse tipo


## **&#9776; ALELOS**

Conforme informações já expostas acima, o gene é uma sequência específica de nucleotídeos no DNA que determina um caráter hereditário. Os genes que determinam uma mesma característica estão situados em um mesmo lugar (locus) em cromossomos homólogos. Os **alelos seriam então formas alternativas de um gene que ocupa o mesmo lócus em cromossomos homólogos**.

Para entendermos melhor, podemos pensar na cor das flores das ervilhas estudadas por Gregor Mendel. Ele percebeu que existiam flores da cor púrpura e flores brancas. Existe, portanto, o gene que determina a cor das flores, e há duas versões (a que determina a cor púrpura e a que determina a cor branca) para esse gene, que são os alelos.

Para cada característica, um **organismo herda dois alelos de um gene, um proveniente da mãe e outro do pai**. Em um indivíduo, os alelos de determinado lócus em cromossomos homólogos podem ser idênticos ou diferentes.

## **&#9776; MALDIÇÃO DA DIMENSIONALIDADE**

A maldição da dimensionalidade, introduzida pela primeira vez pelo matemático Richard Bellman, indica que o número de amostras necessárias para estimar uma função arbitrária com um determinado nível de precisão cresce exponencialmente em relação ao número de variáveis ​​de entrada (isto é, dimensionalidade) da função.

## &#x27A7; **CRIANDO A COLUNA AUSENCIA_GENE**

Conforme estudos realizados a ausência de contagem, presença de valor 0, na tabela ***read count***, nos ofertada, significa que não foi encontrado o gene na leitura do equipamento (ausência de fato do gene, sensibilidade do equipamento baixo,  ou excesso de protocolos). Portanto, como não temos mecanismos de solicitar nova releitura, nos casos de valores ausentes em todos os indivíduos poderemos eliminar essa dada linha. 

Para implementar essa atividade, iremos criar uma coluna denominada [AUSENCIA_GENE] que receberá a soma dos valores da contagem de cada ensamble. 

Caso exista, **o valor 0 na coluna AUSENCIA_GENE**, significa que não há presença desse gene, em nenhum dos indivíduos, o que nos permitirá remover as linhas com esse dado valor.

````python
# criando a coluna AUSENCIA_GENE que recebe o valor de cada indíviduo
# para aquele dado ensamble
df_copia["AUSENCIA_GENE"] = df_copia.sum(axis=1)

# imprime 3 amostras do novo DataFrame
df_copia.head(3)
````

**H1** : Caso não sigam distribuições similares à normal, a hiótese H0 será rejeita e iremos aplicar normalização com o uso de Log.

## &#x27A7; **MÉTODO ELBOW**

Para tentar verficar o K mais otimizado, iremos adotar o método do cotovelo.

````python
# método do cotevelo
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('Método do Cotovelo')
plt.xlabel('Número de Clusters')
plt.ylabel('WCSS')
plt.show()
````
